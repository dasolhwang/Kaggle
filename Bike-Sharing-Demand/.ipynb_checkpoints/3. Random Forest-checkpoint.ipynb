```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor

import matplotlib as mpl
from matplotlib import rc
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.style.use('ggplot')

# 한글폰트 설정
rc('font', family='AppleGothic')
# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
train = pd.read_csv("train.csv", parse_dates=["datetime"])
test = pd.read_csv("test.csv", parse_dates=["datetime"])
print(train.shape, test.shape)
```

## Feature Engineering

```python
def preprocess(data):
    data["year"] = data["datetime"].dt.year
    data["month"] = data["datetime"].dt.month
    data["day"] = data["datetime"].dt.day
    data["hour"] = data["datetime"].dt.hour
    data["dayofweek"] = data["datetime"].dt.dayofweek
    del data["datetime"]
    return data
```

```python
train = preprocess(train)
test = preprocess(test)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(18,5)

plt.sca(ax)
plt.xticks(rotation=30, ha='right')
ax.set(ylabel='렌탈 수',title="훈련데이터의 풍속")
sns.countplot(data=train, x="windspeed", ax=ax)
```

- 풍속이 0 인 데이터가 많다. 잘못입력된 데이터일 가능성 높다.
- 따라서, 풍속을 예측해서 넣어주도록 하자

```python
# 랜덤포레스트를 통해서 풍속 예측하기

def windspeed_predict(data):

    var = ["season", "weather", "humidity", "month", "hour", "temp", "atemp"]
    
    testset = data.loc[data['windspeed'] == 0]
    trainset = data.loc[data['windspeed'] != 0]
    Y = trainset['windspeed']

    rf = RandomForestRegressor()
    rf.fit(trainset[var], Y)
    testset["windspeed"] = rf.predict(X = testset[var])

    data = trainset.append(testset)
    data.reset_index(inplace=True)
    data.drop('index', inplace=True, axis=1)
    
    return data
```

```python
train = windspeed_predict(train) # 분류로 바꿔야겠다!
```
